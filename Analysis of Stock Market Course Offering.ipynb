```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.dropna(subset=['Age', 'Education'], inplace=True)
```

```python
df["Age"]=df['Age'].astype(int)
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x='Mode',data=df)
plt.title('Which Mode has more Customers')
plt.show()
```

```python
df['Age'].unique()
```

```python
a = df['Age']
b = []

for i in a:
    if 12 <= i < 18:
        b.append('12-18')
    elif 18 <= i <= 25:
        b.append('19-25')
    elif 25 < i <= 35:
        b.append('26-35')
    elif 35 < i <= 45:
        b.append('36-45')
    else:
        b.append('NaN')

        
```

```python
df['Age Category']=b
```

```python
df['Age Category'].value_counts()
```

```python
sns.countplot(x='Age Category', data=df, palette='dark')
plt.title('Which Age Category has more interest in Share Market')
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.countplot(x='Institution Name', hue='Experience',data=df)
plt.legend(fontsize=13)
plt.ylabel('Count of Experience', fontsize=12)
plt.xlabel('Institutions',fontsize=14)
plt.title('Customer Experience after Completing the Course from this Institution', fontsize=15)
plt.show()
```
